```python
from price_with_runmean_downloadfunc import download_data_with_runmean
```

```python
d = download_data_with_runmean(companys=['GOOG','VWS.CO', 'TSLA','MAERSK-A.CO'],from_year = 2006)
```

# Bokeh plotting

```python
#imports for plot
from bokeh.io import output_notebook, push_notebook,show
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
from ipywidgets import interact
#This tels python to show the output in this notebook instead of in a html file
output_notebook()
```

```python
# for later reference we make lists of our options
company_list = list(d.columns.levels[0])
```

```python
## interactive legend 

company = company_list[0]
companysource = ColumnDataSource(d[company])
p = figure(x_axis_type='datetime',title=f'Graph with running means for {company}', \
    tools="pan,box_zoom,reset,save", y_axis_label='Closing price', x_axis_label='Date')

p.line(x='Date', y='Close', source=companysource, color = 'blue', legend= 'Closing price',muted_alpha=0.2)
p.line(x='Date', y='rm_200', source=companysource, color='black', line_dash='4 4', legend = '200 days',muted_alpha=0.2)
p.line(x='Date', y='rm_50', source=companysource, color='red', line_dash='4 4',legend = '50 days',muted_alpha=0.2)
p.legend.location = "top_left"
p.legend.click_policy="mute"

def update_name(company):
    p.title.text = 'Closing price and running mean of ' + str(company)
    companysource.data = ColumnDataSource(d[str(company)]).data
    push_notebook()
    
show(p,notebook_handle=True)


interact(update_name,company=company_list)
```
